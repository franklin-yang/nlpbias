```python
# fname = 'doubly_eponymous_hard.jsonl_gender_balanced'
# fname = 'doubly_eponymous_easy.json_gender_balanced'
fname='doubly_eponymous_cranfield.jsonl_gender_balanced scores'
import json
import math

from collections import Counter
pairs = Counter()
ps = []
idx = 0
rows = []
with open(fname) as file:
    lines = file.readlines()
    print(len(lines))
    for line in lines:
        row = []
        tokens = line.split(',')
        if len(tokens) < 3:
            # header or bad format
            continue
#         print(",".join(tokens[2:]))
        demographic_data = json.loads(",".join(tokens[2:]).replace("'",'"'))
        try:
            row.append(math.log(float(tokens[0])/float(tokens[1])))
            maleFound = femaleFound = False
            for demographics in demographic_data:
                if demographics['gender'] == 'male':
                    maleDemo = demographics
                elif demographics['gender'] == 'female':
                    femaleDemo = demographics
            row.append(maleDemo['nationality'])
            row.append(femaleDemo['nationality'])
#             if 'unknown' in row:
#                 row = []
#                 continue
            rows.append(row)
                
#        
#             print(f"question: {example['question']}, \n model answer: {model_ans}, \n ground truth: {example['options']}\n accuracy: {acc}")
        except Exception as e:
            raise e
            print(e)
            pass
        idx += 1
```

```python
from numpy import asarray
from sklearn.preprocessing import OrdinalEncoder
# define data
from sklearn.preprocessing import OneHotEncoder

data = asarray(rows)
encoder = OneHotEncoder(sparse=False)
# transform data
onehot = encoder.fit_transform(data[:,1:])
X=onehot
y=data[:,:1]
print(onehot)
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(onehot, data[:,:1])
# print(data)
```

```python
reg.score(X,y)
```
